```python
class Graph:
    def __init__(self, gdict=None):
        if gdict is None:
            gdict = {}
        self.gdict = gdict

    def getVertices(self):
        return list(self.gdict.keys())

    def edges(self):
        return self.findedges()

    def findedges(self):
        edgename = []
        for vrtx in self.gdict:
            for nxtvrtx in self.gdict[vrtx]:
                if (nxtvrtx, vrtx) not in edgename:
                    edgename.append((vrtx, nxtvrtx))
        return edgename

# Graph TUV
graph_TUV = {
    "T": ["U", "W"],
    "U": ["T", "V", "X"],
    "V": ["U"],
    "W": ["T", "X", "Z"],
    "X": ["U", "W", "S"],
    "Z": ["W"],
    "S": ["X"]
}

g = Graph(graph_TUV)
edges_tuv = g.edges()
print("Edge Graph TUV:", edges_tuv)
print("Daftar Edge:")
for i, edge in enumerate(edges_tuv, 1):
    print(f"{i}. {edge[0]} - {edge[1]}")
print("Jumlah Edge:", len(edges_tuv))
```

LATIHAN 9 (HALAMAN 42)

```python
class Graph:
    def __init__(self, gdict=None):
        if gdict is None:
            gdict = {}
        self.gdict = gdict

    def getVertices(self):
        return list(self.gdict.keys())

    def edges(self):
        return self.findedges()

    def findedges(self):
        edgename = []
        for vrtx in self.gdict:
            for nxtvrtx in self.gdict[vrtx]:
                if (nxtvrtx, vrtx) not in edgename:
                    edgename.append((vrtx, nxtvrtx))
        return edgename

    def addVertex(self, vrtx):
        if vrtx not in self.gdict:
            self.gdict[vrtx] = []

    def AddEdge(self, edge):
        edge = set(edge)
        (vrtx1, vrtx2) = tuple(edge)
        if vrtx1 in self.gdict:
            if vrtx2 not in self.gdict[vrtx1]:
                self.gdict[vrtx1].append(vrtx2)
        else:
            self.gdict[vrtx1] = [vrtx2]

        # Untuk graph tidak berarah, tambahkan juga sebaliknya
        if vrtx2 in self.gdict:
            if vrtx1 not in self.gdict[vrtx2]:
                self.gdict[vrtx2].append(vrtx1)
        else:
            self.gdict[vrtx2] = [vrtx1]

# Graph TUV
graph_TUV = {
    "T": ["U", "W"],
    "U": ["T", "V", "X"],
    "V": ["U"],
    "W": ["T", "X", "Z"],
    "X": ["U", "W", "S"],
    "Z": ["W"],
    "S": ["X"]
}

g = Graph(graph_TUV)
print("Edge sebelum:", g.edges())
g.AddEdge({'T', 'Z'})  # T ke Z
g.AddEdge({'V', 'S'})  # V ke S
print("Edge setelah:", g.edges())
print("Graph terupdate:")
for vertex, neighbors in g.gdict.items():
    print(f"  {vertex}: {neighbors}")
```

```python
# Contoh graph berbobot untuk MST dan Shortest Path
print("\n1. MINIMUM SPANNING TREE (MST)")

# Graph contoh dengan bobot
graph_weighted = {
    'A': [('B', 5), ('D', 2)],
    'B': [('A', 5), ('C', 3), ('E', 4)],
    'C': [('B', 3), ('F', 1), ('E', 7)],
    'D': [('A', 2), ('E', 6)],
    'E': [('B', 4), ('C', 7), ('D', 6), ('F', 7)],
    'F': [('C', 1), ('E', 7)]
}

print("Graph berbobot:")
for node, neighbors in graph_weighted.items():
    print(f"  {node}: {neighbors}")

print("\nALGORITMA PRIM (mulai dari A):")
prim_steps = [
    "1. Pilih edge A-D (bobot 2)",
    "2. Pilih edge A-B (bobot 5)",
    "3. Pilih edge B-C (bobot 3)",
    "4. Pilih edge C-F (bobot 1)",
    "5. Pilih edge B-E (bobot 4)"
]
for step in prim_steps:
    print(f"   {step}")

print("   Total bobot MST: 2 + 5 + 3 + 1 + 4 = 15")

print("\nALGORITMA KRUSKAL:")
kruskal_steps = [
    "1. Pilih edge C-F (bobot 1)",
    "2. Pilih edge A-D (bobot 2)",
    "3. Pilih edge B-C (bobot 3)",
    "4. Pilih edge B-E (bobot 4)",
    "5. Pilih edge A-B (bobot 5)"
]
for step in kruskal_steps:
    print(f"   {step}")

print("   Total bobot MST: 1 + 2 + 3 + 4 + 5 = 15")

print("\n2. SHORTEST PATH (A ke F)")
print("METODE GREEDY:")
greedy_path = ["A → B (5)", "B → C (3)", "C → F (1)"]
print("   Jalur: " + " → ".join(greedy_path))
print("   Total jarak: 5 + 3 + 1 = 9")

print("\nJALUR OPTIMAL:")
optimal_path = ["A → D (2)", "D → E (6)", "E → F (7)"]
print("   Jalur: " + " → ".join(optimal_path))
print("   Total jarak: 2 + 6 + 7 = 15")

print("\n3. MENGAPA GREEDY KADANG GAGAL:")
reasons = [
    "• Memilih solusi terbaik lokal tanpa mempertimbangkan konsekuensi global",
    "• Dapat terjebak dalam minimum lokal",
    "• Tidak melihat gambaran keseluruhan path",
    "• Contoh: greedy memilih B (5) daripada D (2) karena terlihat lebih dekat ke C"
]
for reason in reasons:
    print(f"   {reason}")
```
