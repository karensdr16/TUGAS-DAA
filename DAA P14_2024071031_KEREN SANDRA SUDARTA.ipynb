```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

vertices = range(1,10)
edges = [(7,2), (2,3), (7,4), (4,5), (7,3),(7,5), (1,6), (1,7), (2,8), (2,9)]
G.add_nodes_from(vertices)
G.add_edges_from(edges) # Changed from G.add_nodes_from(edges)

degree_values = dict(G.degree())

dos_values = {node : degree for node, degree in degree_values.items()}

print("DOS values:")
for node, dos in dos_values.items():
  print(f"{node}: {dos}")

# Check if max_degree is zero before calculating normalized_dos_values
max_degree = max(degree_values.values())
if max_degree == 0:
  print("\nCannot normalize DOS values: All node degrees are zero.")
else:
  normalized_dos_values = {node: round((degree / max_degree) * 9, 2) for node, degree in degree_values.items()}

  print("\nNormalized DOS values:")
  for node, normalized_dos in normalized_dos_values.items():
    print(f"{node}: {normalized_dos}") # Changed from {node}: {dos}
```

```python
## Latihan 1

import networkx as nx
import matplotlib.pyplot as plt

# Buat graf
G = nx.Graph()

# Tambah node
nodes = range(1, 11)
G.add_nodes_from(nodes)

# Tambah edge (hubungan antar node) sesuai gambar
edges = [
    (1, 2), (1, 7),
    (2, 3), (2, 7),
    (3, 4), (3, 6),
    (5, 6),
    (6, 7),
    (7, 8), (7, 9), (7, 10),
    (8, 9)
]
G.add_edges_from(edges)

# Posisi tetap (x, y) untuk tiap node
pos = {
    1: (0.0, 2.0),
    2: (0.0, 1.0),
    3: (0.5, 0.0),
    4: (-0.5, -0.8),
    5: (1.0, -0.8),
    6: (1.0, 0.3),
    7: (1.0, 1.5),
    8: (2.0, 2.0),
    9: (3.0, 2.0),
    10: (2.0, 1.0)
}

# Label F dan NF di dalam node
labels_inside = {
    1: "F",
    2: "NF",
    3: "NF",
    4: "F",
    5: "F",
    6: "NF",
    7: "F",
    8: "NF",
    9: "NF",
    10: "F"
}

# Gambar node dan edge
plt.figure(figsize=(6, 6))
nx.draw_networkx_edges(G, pos, width=2)

# Node kosong (hanya lingkaran)
nx.draw_networkx_nodes(
    G, pos,
    node_size=1500,
    node_color="white",
    edgecolors="black"
)

# Nomor node di luar (atas)
node_number_pos = {n: (x, y + 0.22) for n, (x, y) in pos.items()}
nx.draw_networkx_labels(G, node_number_pos,
                        labels={n: str(n) for n in G.nodes()},
                        font_size=10)

# Label F / NF di dalam node
nx.draw_networkx_labels(G, pos,
                        labels=labels_inside,
                        font_size=10)

plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

# Node 113
G.add_nodes_from(range(1, 14))

# Edge sesuai gambar
edges = [
    (1, 2), (1, 3),
    (2, 3), (2, 12),
    (3, 4), (3, 5), (3, 10), (3, 11), (3, 12),
    (4, 5), (4, 6), (4, 7),
    (5, 6), (5, 7), (5, 8),
    (6, 9), (6, 5),
    (7, 8),
    (8, 9),
    (10, 11),
    (11, 12),
    (12, 13)
]
G.add_edges_from(edges)

# Perkiraan posisi tetap (silakan sedikit disesuaikan kalau mau lebih presisi)
pos = {
    1: (0.0, 2.0),
    2: (0.5, 1.2),
    3: (1.2, 1.4),
    4: (2.4, 2.0),
    5: (3.0, 1.4),
    6: (2.8, 0.6),
    7: (3.6, 2.0),
    8: (4.2, 1.2),
    9: (3.8, 0.3),
    10: (1.8, -0.1),
    11: (1.4, 0.4),
    12: (0.8, 0.4),
    13: (0.2, -0.3)
}

# Label F / NF (sesuai gambar)
labels_inside = {
    1: "F",
    2: "F",
    3: "F",
    4: "NF",
    5: "NF",
    6: "NF",
    7: "NF",
    8: "NF",
    9: "NF",
    10: "F",
    11: "F",
    12: "F",
    13: "NF"
}

# Warna: F = merah, NF = hijau
color_map = []
for n in G.nodes():
    if labels_inside[n] == "F":
        color_map.append("red")
    else:
        color_map.append("green")

plt.figure(figsize=(8, 4))
nx.draw_networkx_edges(G, pos, width=2)

nx.draw_networkx_nodes(
    G, pos,
    node_size=1500,
    node_color=color_map,
    edgecolors="black"
)

# Nomor node di atas lingkaran
node_number_pos = {n: (x, y + 0.25) for n, (x, y) in pos.items()}
nx.draw_networkx_labels(
    G,
    node_number_pos,
    labels={n: str(n) for n in G.nodes()},
    font_size=9
)

# Label F / NF di dalam node
nx.draw_networkx_labels(
    G,
    pos,
    labels=labels_inside,
    font_size=9,
    font_color="black"
)

plt.axis("off")
plt.tight_layout()
plt.show()
```
